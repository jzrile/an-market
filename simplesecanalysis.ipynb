```python
# ffn is Financial Functions for Python library
# https://pmorissette.github.io/ffn/index.html
# !pip install ffn

import ffn

import matplotlib.pyplot as plt

%matplotlib inline 
%pylab inline 
pylab.rcParams['figure.figsize'] = (15, 9)

# Import prices from yahoo finance, set period. 
# Here are the biggest sp500 etfs (SPDR SPY, Vanguard voo,ishares ivv and ishares iuse - EUR hedged sp500 etf)
prices = ffn.get('spy,voo,ivv,iuse.mi',  start='2016-01-01')
prices.head()




```

```python
#plot time series
ts = prices.rebase().plot()
```

```python
# Returns of each security and plot histogram 
returns = prices.to_returns().dropna()
# ts = returns.hist(figsize(15, 8))
```

```python
# plot correlation map
returns.plot_corr_heatmap()
```

```python
#prints table with performance and risk data

perf = prices.calc_stats()
perf.set_riskfree_rate(0.01)
perf.display();
```
